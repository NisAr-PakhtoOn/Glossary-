```python
# converting those categorical features into numerical features
county_name = pd.get_dummies(df['county_name'])
county_name.head()
```

```python
state = pd.get_dummies(df['state'])
state.head()
```

```python
urbanicity = pd.get_dummies(df['urbanicity'])
urbanicity.head()
```

```python
political_affiliation = pd.get_dummies(df['political_affiliation'])
political_affiliation.head()
```

```python
net_worth = pd.get_dummies(df['net_worth'])
net_worth.head()
```

```python
marital_status = pd.get_dummies(df['marital_status'])
marital_status.head()
```

```python
investment_type = pd.get_dummies(df['investment_type'])
investment_type.head()
```

```python
family_position = pd.get_dummies(df['family_position'])
family_position.head()
```

```python
gender = pd.get_dummies(df['gender'])
gender.head()
```

```python
# merging the dataset again
merged = pd.concat([county_name,state,political_affiliation,net_worth,marital_status,investment_type,family_position,gender], axis = 1)
merged.head()
```

```python
# dimensionality reduction using PCA
```

```python
from sklearn.decomposition import PCA
```

```python
# we are passing component as 10, we want to reduce the dimensions from 675 to just 10 features
pca = PCA(n_components=10)
```

```python
pca.fit(merged)
```

```python
x_pca = pca.transform(merged)
```

```python
x_pca.shape
```

```python
x_pca
```

```python
df1 = pd.DataFrame(x_pca)
```

```python
# final dataframe with diemnsionality reduced to 10
df1
```

```python
# plotting the correlation graph again
corr = df1.corr()
print(corr)
sns.set(rc = {'figure.figsize':(10,5)})
sns.heatmap(corr,xticklabels=corr.columns,yticklabels=corr.columns)
plt
```
